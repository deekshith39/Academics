## Multiple Linear Regression

### Dataset:

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"D:\6th Sem\Machine Learning\LAB CIE 2\dataset.csv")
```

```python
df.head()
```

### Splitting the dataset into test and train

```python
X = df[['Age', 'Height']]
y = df[['Weight']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

### Training

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

### Visualizations

```python
#Visualization
reg = LinearRegression()
reg.fit(X_train[['Age']], y_train.Weight)
plt.scatter(X_train.Age,y_train.Weight,s=40)
plt.plot(X_train.Age, reg.predict(X_train[['Age']]), color='red')
plt.xlabel('Age')
plt.ylabel('Weight')
```

```python
#Visualization
reg1 = LinearRegression()
reg1.fit(X_train[['Height']], y_train.Weight)
plt.scatter(X_train.Height,y_train.Weight,s=40)
plt.plot(X_train.Height, reg1.predict(X_train[['Height']]), color='red')
plt.xlabel('Height')
plt.ylabel('Weight')
```

```python
#Visualization
reg2 = LinearRegression()
reg2.fit(X_test[['Age']], y_test.Weight)
plt.scatter(X_test.Age,y_test.Weight,s=40)
plt.plot(X_test.Age, reg2.predict(X_test[['Age']]), color='red')
plt.xlabel('Age')
plt.ylabel('Weight')
```

```python
#Visualization
reg3 = LinearRegression()
reg3.fit(X_test[['Height']], y_test.Weight)
plt.scatter(X_test.Height,y_test.Weight,s=40)
plt.plot(X_test.Height, reg3.predict(X_test[['Height']]), color='red')
plt.xlabel('Height')
plt.ylabel('Weight')
```

### Prediction

Question : Height = 5.5, Age = 38, Weight = ?

```python
model.predict([[38, 5.5]])
```

Math behind it :

```python
#y=m1*x1+m2*x2+b
```

Slopes:

```python
model.coef_
```

Intercept:

```python
model.intercept_
```

```python
y = model.coef_[:,0]*38 + model.coef_[:,1]*5.5 + model.intercept_
y
```

### Accuracy

```python
model.score(X_test, model.predict(X_test))
```
